```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np


x =[12,23,45,12,23,36,25,1,4,26,45,65,48,75,62,98,32]
y =[11,2,5,32,26,16,28,19,49,27,85,95,28,75,82,91,20]

data = list(zip(x,y))

dbscan = DBSCAN(eps=10, min_samples=2)


labels = dbscan.fit_predict(data)

print("result for the prediction :")
plt.figure(figsize=(8,5))
plt.scatter(x,y,c=labels,cmap="plasma",s=115)
plt.title("plot for the DBSCAN clustering")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate synthetic dataset with 3 blobs
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)  # tune eps & min_samples
labels = dbscan.fit_predict(X)
print(labels)
# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", s=50)
plt.title("DBSCAN Clustering on Blobs Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

```
