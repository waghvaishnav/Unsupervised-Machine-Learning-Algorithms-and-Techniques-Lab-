#K - Means CLUSTERING :

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data= {"customer":["riya","aman","shreya","imran","umesh","rahul"],
       "age":[20,30,40,22,38,25],
       "spending":[100,200,300,110,290,130]
}

df = pd.DataFrame(data)
x= df[["age","spending"]]

model = KMeans(n_clusters=2,random_state=45,n_init=10)

df["group"] = model.fit_predict(x)
print(df['group'])
plt.figure(figsize=(8,7))

for group in df["group"].unique():
  print(group)
  group_data = df[df["group"]== group]
  print(group_data)
  plt.scatter(group_data["age"],group_data["spending"],label=f'group {group}')

plt.xlabel("age")
plt.ylabel("spending ",rotation =45)
plt.grid(True)
plt.title("scatter plot for the age and spending")
plt.legend()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data= {"customer":["riya","aman","shreya","imran","umesh","rahul"],
       "age":[20,30,40,22,38,25],
       "spending":[100,200,300,110,290,130]
}

df = pd.DataFrame(data)

x = df[['age',"spending"]]

model = KMeans(n_clusters=2,random_state=42,n_init=10)

df["group"] = model.fit_predict(x)

for group in df['group'].unique():
  group_datas = df[df["group"] == group]
  plt.scatter(group_datas['age'],group_datas["spending"],label= f'group{group}')

plt.xlabel("age")
plt.ylabel("spending ",rotation =45)
plt.grid(True)
plt.title("scatter plot for the age and spending")
plt.legend()
```
